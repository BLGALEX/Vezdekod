```python
import numpy as np
import re
import matplotlib.pyplot as plt
```

```python
text
with open("data.txt", "r") as f:
    text = f.read()
    print(text[2:-2])
```

```python
numbers = re.findall('[0-9]+\.*[0-9]*', text)
```

```python
phi = []
r = []
max_dist = 0
for i in range(len(numbers)//3):
    dist = float(numbers[i*3+2])
    if dist < 5:
        dist = 5
    elif dist > 4000:
        dist = 4000
    if dist > max_dist:
        max_dist = dist
    phi.append((float(numbers[i*3+1])*np.pi)/180)
    r.append(dist)
```

```python
fig = plt.figure(figsize=(10, 10))
ax = plt.subplot(projection='polar')
ax.plot(phi, r, 'o', color='b', markersize=max_dist/1000)
ax.set_rticks([0, max_dist//5, (max_dist//5)*2, (max_dist//5)*3, (max_dist//5)*4, max_dist])  
ax.set_rlabel_position(5)
ax.grid(True)
ax.set_title("Laser", va='bottom')
plt.show()
```
